# Problem-1 : Write a code for 1-1 dot plot

```python
import numpy as np
import random as rn
import matplotlib.pyplot as plt
import string
```

```python
seq1 = "UAACUGGGCGUUAGCGAGCAAUGAC"
seq2 = "CCCGCGCAUCGCACUGGACUGUACU" 
```

```python
seq_list1 = list(seq1)
seq_list2 = list(seq2)
```

```python
seq_list1
```

```python
x=[]
y=[]
for i in range(len(seq_list1)):
    if(seq_list1[i]==seq_list2[i]):
        x.append(i)
```

```python
#figure = plt.figure(figsize=(8,8))

plt.rcParams["figure.figsize"] = (9,9)
plt.scatter(x,x)

xt=plt.xticks(np.arange(len(list(seq1))),list(seq1))
yt=plt.yticks(np.arange(len(list(seq2))),list(seq2))


plt.xlabel("Sequance 1",fontsize=20)
plt.ylabel("Sequance 2",fontsize=20)
plt.title("One-One Dot Plot",fontsize=30)
plt.grid()
plt.show()
```

# Problem-2 : Write a code for sliding window

```python
def Sliding_window(seq1,seq2,thresold,window):
    seq_list1=list(seq1)
    seq_list2=list(seq2)

    seq11 = ["".join(seq_list1[i:i+window]) for i in range(0, len(seq_list1)) if(len(seq_list1[i:i+window])==window)] 
    seq22 = ["".join(seq_list2[i:i+window]) for i in range(0, len(seq_list2)) if(len(seq_list2[i:i+window])==window)]
    
    x=[]
    y=[]
    for i in range(len(seq11)):    
        for j in range(len(seq22)):
            count=0    
            for k in range(window):
                if(seq11[i][k]==seq22[j][k]):
                    count+=1
            if(count>=thresold):
                x.append(seq11.index(seq11[i]))
                y.append(seq22.index(seq22[j]))
                
    plt.figure(figsize=(20,20))
    plt.scatter(x,y)

    xt=plt.xticks(np.arange(len(list(seq11))),list(seq11),rotation=90)
    yt=plt.yticks(np.arange(len(list(seq22))),list(seq22))

    plt.xlabel("Sequance 1 ({})".format(seq2),fontsize=20)
    plt.ylabel("Sequance 2 ({})".format(seq1),fontsize=20)
    plt.title("Window size = {}, thresold={}".format(window,thresold),fontsize=30)
    plt.grid()            
   
```

```python
Sliding_window(seq1,seq2,thresold=10,window=10)
```
