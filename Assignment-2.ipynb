Problem:1: Write a code for calculating the frequency of each nucleotide  ('nt'). and then draw the histograms.

```python
import random
nt = ["A","T","G","C"]
```

```python
# dna sequence creation

def DNA(nt):
    sequences =[]
    
    for i in range(100):
        dna =[]
        for j in range(500):
            l= random.choice(nt)
            dna.append(l)
        a="".join(dna) 
        sequences.append(a)
    return sequences    
            
```

```python
sequences = DNA(nt)
print(sequences)
```

frequency of each nucleotide ('nt').

```python
# adding frequency in terms of dictionary
freq = {
    'A':0,
    "T":0,
    "G":0,
    "C":0,
}
for i in sequences:
    for j in freq:
        freq[j]+=i.count(j)
```

```python
print(freq)
```

```python
import matplotlib.pyplot as plt
plt.bar(*zip(*freq.items()))
```
