Problem -2. Create 100 random string with the four alphabets A,T,G and C of length 50 using biopython



```python
lst1 = list(["A","T","G","C"])

String1 = []
for i in range(100):
    temp = "".join(random.choice(lst1) for i in range(1,51))
    String1.append(temp)
    
```
